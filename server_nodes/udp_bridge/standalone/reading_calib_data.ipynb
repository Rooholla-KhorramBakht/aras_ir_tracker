```python
import pickle 
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
with open('recorded_data/recorded-2021-06-07-00:29:40.pckl','rb') as f:
    data = pickle.load(f)
print('The Port ID of the stations are:')
print(data.keys())
print('A random entry in this data structure is: time, markers')
print(data[5000][0])
print(data[5001][0])
print(data[5003][0])
```

```python
station_5000_cam2 = np.vstack([marker[1][1] for marker in data[5000]])
station_5000_cam1 = np.vstack([marker[1][0] for marker in data[5000]])
station_5000_ts = np.vstack([marker[0] for marker in data[5000]])

station_5001_cam2 = np.vstack([marker[1][1] for marker in data[5001]])
station_5001_cam1 = np.vstack([marker[1][0] for marker in data[5001]])
station_5001_ts = np.vstack([marker[0] for marker in data[5001]])

station_5003_cam2 = np.vstack([marker[1][1] for marker in data[5003]])
station_5003_cam1 = np.vstack([marker[1][0] for marker in data[5003]])
station_5003_ts = np.vstack([marker[0] for marker in data[5003]])
```

```python
plt.plot(station_5000_ts)
plt.plot(station_5001_ts)
plt.plot(station_5003_ts)
```

```python
s5000_cam2_idx = np.where(station_5000_cam2[:,0]!=-1)[0]
s5000_cam1_idx = np.where(station_5000_cam1[:,0]!=-1)[0]

station_5000_cam2_with_markers=\
    station_5000_cam2[s5000_cam2_idx,:]

station_5000_cam1_with_markers=\
    station_5000_cam1[s5000_cam1_idx,:]

plt.plot(station_5000_cam2_with_markers[:,0], 
         station_5000_cam2_with_markers[:,1])

plt.plot(station_5000_cam1_with_markers[:,0], 
         station_5000_cam1_with_markers[:,1])
```
